```python
%matplotlib inline
from toy_plot_helpers import ToyPlot
plot = ToyPlot()
```

```python
import openpathsampling as paths
import numpy as np
```

## Setting up the system

```python
import openpathsampling.toy_dynamics as toys
pes = (toys.toy_pes.OuterWalls([1.0, 1.0], [0.0, 0.0]) + 
       toys.toy_pes.Gaussian(2.0, [1.0, 4.0], [0.0,1.0]) + 
       toys.toy_pes.Gaussian(2.0, [25.0, 1.0], [0.0, -1.0])
      )
topology=paths.ToyTopology(
    n_spatial = 2,
    masses =[1.0, 1.0],
    pes = pes
)
template_snap = paths.Snapshot(
    coordinates=np.array([[0.0, 0.0]]), 
    velocities=np.array([[0.0,0.0]]),
    potential_energy = 0.0, # to set dimensionless energies for toy engine
    kinetic_energy = 0.0,
    topology=topology
)
integ = toys.toy_integrators.LangevinBAOABIntegrator(dt=0.02, temperature=0.1, gamma=1.0)
options={'integ' : integ,
         'n_frames_max' : 5000,
         'nsteps_per_frame' : 10
        }
sim = toys.toy_engine.ToyEngine(
        options=options,
        template=template_snap
)

paths.PathMover.engine = sim
sim.initialized = True
```

## Load a trajectory

```python
storage = paths.storage.Storage('bootstrap.nc', mode='r')
traj = storage.sampleset[len(storage.sampleset)-1].samples[-1].trajectory
```

```python
plot.add_pes(pes)
```

```python
plot.plot_pes_initcond([traj])
```

## Defining states and interfaces

```python
collectivevariable = paths.CV_Function(name="xval",
                                   fcn=lambda snap : snap.xyz[0,0])
```

```python
stateA = paths.LambdaVolume(collectivevariable, float("-inf"), -0.3)
stateB = paths.LambdaVolume(collectivevariable, 0.3, float("inf"))
interfaces = paths.VolumeFactory.LambdaVolumeSet(collectivevariable, 
                                                 minvals=float("-inf"), 
                                                 maxvals=[-0.30, -0.25, -0.20, -0.13, -0.05, 0.0]
                                                )
```

```python
plot.add_states([stateA, stateB])
plot.add_interfaces(interfaces)
```

```python
plot.plot([traj])
```

## Automatic preparation of TIS pathmovers and ensembles

```python
tis = paths.analysis.TISTransition(stateA, stateB, collectivevariable, interfaces)
```

```python
tis.ensembles
```

```python
tis.movers
```

## Set up and run a TIS simulation

```python
alltrajs = paths.SampleSet(
    [paths.Sample.initial_sample(replica=tis.ensembles.index(e), 
                                 trajectory=paths.Trajectory(traj), # makes a copy
                                 ensemble=e)
     for e in tis.ensembles]
)
```

```python
tis_calc = paths.PathSampling(
    storage=None,
    engine=sim,
    root_mover=paths.SequentialMover(movers=tis.movers['shooting']),
    globalstate=alltrajs
)
```

```python
tis_calc.run(20)
```

```python
plot.plot([sample.trajectory for sample in tis_calc.globalstate])
```
